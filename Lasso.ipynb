```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor
```

```python
from sklearn.model_selection import cross_val_score
mse=cross_val_score(regressor,x,y,scoring='neg_mean_squared_error', cv=5)
mse
```

```python
np.mean(mse)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
```

```python
parameters={'alpha':[1e-15,1e-10,1e-8,1e-5,1e-2,1,5,10,20,30,40,50]}
redge_regressor = GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5,verbose=1)
```

```python
redge_regressor.fit(x,y)
```

```python
redge_regressor.best_params_
```

```python
redge_regressor.best_score_
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
pre_ridge=redge_regressor.predict(x_test)
```

```python
import seaborn as sns
```
