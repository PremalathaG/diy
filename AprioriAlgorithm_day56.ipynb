```python

import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_csv('Groceries data.csv')
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
df[df.Member_number == 1001] 
```

```python
df["item_count"] = 1
```

```python
df = df.drop(columns = ["month","day","year","day_of_week"])
```

```python
df.shape
```

```python
df.Member_number.value_counts()
```

```python
basket = (df.groupby(['Member_number','itemDescription'])['item_count']
          .sum()
          .unstack()
          .reset_index()
          .fillna(0)
          .set_index('Member_number'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0    
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
basket_sets.head()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)


rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)

rules.head()
```

```python
rules[ (rules['lift'] >= 1 ) & (rules['confidence'] >= 0.5) ].head()
```
